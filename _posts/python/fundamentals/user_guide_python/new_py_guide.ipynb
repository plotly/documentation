# Python API User Guide

### What is Plotly?

Plotly at its core is a data visualization toolbox. Under every plotly graph is a JSON object, which is a dictionary like data structure. By changing the values of some keywords in this object, we can get vastly different and ever more detailed plots. Let's start with a basic example to better understand how plots are created with the Plotly Python Library. We'll go into more detail regarding the different parts as we move forward.

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Scatter(x = [1,2,3], y = [4,5,6], marker = {'color': '#CAA5F7', 'symbol': 104, 'size': 10}, 
                    mode = "markers+lines",  text = ["one","two","three"], name = '1st Trace')

layout = go.Layout(title = "First Plot", xaxis = {'title':'x1'}, yaxis = {'title':'x2'}, showlegend=True)
data=[trace1]
figure = go.Figure(data, layout)
py.iplot(figure, filename = 'pyguide_1')
```

### The Source of Plotly's Power

All the graphs and plots which Plotly generates are actually the product of our javascript library [`plotly.js`](https://plot.ly/javascript). Whether you see Plotly graphs in a browser, or an IPython notebook, all the visualizations and interactiveness are made possible by [`plotly.js`](https://plot.ly/javascript). Built on top of `d3.js` and `stack.gl`, [`plotly.js`](https://plot.ly/javascript) is a high-level, declarative charting library. [`plotly.js`](https://plot.ly/javascript) ships with over 30 chart types, including 3D charts, statistical graphs, and SVG maps.

This way, the package will provide functions and graph objects that will simplify the process of generating plots, which would amount to properly defining keywords in a JSON object (as seen above). To this end, the Python API provides functions and graph objects which will allow us to create our plots functionally. So let's break this down.

### Data 
Data is a list object in Python, which contains all the traces that you want to plot. 'Trace' is a name we give to a collection of data and the specifications of which we want that data plot. Notice that a trace will also be an object itself, and these will be named according to how you want the data displayed on the plotting surface. 

```python
data
```

Plotly plots are interactive, meaning you can manually explore the data by panning, selecting, zooming on the graphing surface (among other possible actions, try panning the axes!). To customize your plot see **[possible keywords](https://plot.ly/python/reference/#scatter)** associated with scatter plots.

In no time you will be able to figure out how to create plots the way you want them, and then share them with others. Let's take a quick look at how we would define the required objects for generatting a scatterplot comparing life expectancies and GDP per capital between two different continents.  

```python
import plotly.plotly as py
import plotly.graph_objs as go
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/yankev/test/master/life-expectancy-per-GDP-2007.csv')

americas = df[(df.continent=='Americas')]
europe = df[(df.continent=='Europe')]

trace0 = go.Scatter(
    x=americas.gdp_percap, y=americas.life_exp,
    mode='markers', marker={'size': 20,
    'line':{'width':1}, 'color':"#F36886"},
    name='U.S.', text=americas.country)

trace1 = go.Scatter(
    x=europe.gdp_percap, y=europe.life_exp,
    mode='markers', marker={'size':20, 'line':{'width':1},
    'color':"#28C3D4"}, name='Europe', text=europe.country)

layout = go.Layout(
    title='Life Expectancy & GDP per Capital in 2007',
    hovermode='closest', xaxis={'title':'GDP per capita (2000 dollars)',
    'ticklen':5, 'zeroline':False, 'gridwidth':2},
    yaxis={'title':'Life Expectancy (years)', 'ticklen':5, 'gridwidth':2})
    
py.iplot(go.Figure([trace0, trace1], layout), filename='life-expectancy-per-GDP-2007')
```

```python
go.Scatter(x=[1,2,3], y=[4,5,6], marker={'color':'red', 'symbol':104, 'size':10}, mode="markers+lines",  text=["one", "two", "three"])
```

`go.Scatter` defines a trace producing a scatter plot. It also defines the data that we want to be plotted, which are 3 data points (1,4), (2,5), (3,6), as well as a miriad of specifications related to plotting this data. In this example we wanted the points to be plotted as hollow x's with lines joining them, all in red. To add another Scatter object to our data list, we need to define a new Scatter object, and include it inside our data object.
#### First let's make up some cool data to plot:

```python
import numpy as np
x = np.arange(1,3.2,0.2)
y = 6*np.sin(x)
y
```

```python
trace1 = go.Scatter(x = [1,2,3], y = [4,5,6], marker = {'color': '#CAA5F7', 'symbol': 104, 'size': 10}, mode = "markers+lines", text = ["one","two","three"], name = '1st Trace')
trace2 = go.Scatter(x=x, y=y, marker={'color': 'blue', 'symbol': 'star', 'size': 10}, mode='markers', name='2nd trace')
data = go.Data([trace1, trace2])
data
```

```python
plot2 = py.iplot(go.Figure(data, layout), filename='pyguide')
plot2 
```

#### Layout

The Layout object is about features for styling the plot and will define how the plot would look like. Changing title, axis titles, spacing, font and even draw shapes on top of the plot would all be performed in Layout. 

##### Annotations

We added a plot title as well as titles for all the axes. 
For fun we could add some text annotation as well in order to indicate the maximum point that's been plotted on the current plotting surface.

```python
layout.update(dict(annotations=[go.Annotation(text="Highest Point", x=3, y=6)]))
py.iplot(go.Figure(data, layout), filename='pyguide_1')
```

##### Shapes

Let's add a rectangular block to highlight the section where trace 1 is above trace2.

```python
layout.update({'shapes': [
        # 1st highlight during Feb 4 - Feb 6
        {'type': 'rect',
            # x-reference is assigned to the x-values
            'xref': 'x',
            # y-reference is assigned to the plot paper [0,1]
            'yref': 'y', 'x0': '1', 'y0': 0,
            'x1': '2', 'y1': 7, 'fillcolor': '#d3d3d3',
            'opacity': 0.2, 'line': {'width': 0}
        }]})

py.iplot(go.Figure(data, layout), filename='stars')
```

Of course, all this can be found in the `Layout` section of the [Reference Page](https://www.plot.ly/python/referece).

#### Figure

Finally, we get to the figure object. Figure as an input of `py.iplot` is a dictionary-like object:

```python
figure
```

`go.Figure` just creates the final object to be plotted, and simply just creates a dictionary-like object that contains both the data object and the layout object.

```python
go.Figure(data, layout)
```

#### Why `graph_objs`?

After viewing the outputs of these functions (ie: the objects), we can see that they are just lists or dictionaries. But they're a little more than that. Though they do inherit properties from dictionaries (traces, and layout) and lists (figure), they provide a bit more functionality as we'll soon see. Not to mention the fact that it's much simpler to create plots in this functional fashion compared to manually writing up a dictionary. 

The first neat option about using graph_objs is that you can call help on them.

```python
help(go.Figure)
```

```python
help(go.Scatter)
```

As you can see, calling help shows us all the attributes that the Scatter object takes as parameters. Also because the scatter object is based off of a dictionary, you can see all the different methods that are attached to this object as well. For example, let's see the use of `update` method:

```python
figure.update(layout = {'title':'Updated Title'}, data = [{'marker':{'color':'#23EAE6'}}])
py.iplot(figure, filename = 'updatePlot')
```

Focussing more on the parameters, we can see that the objects have key-word/name validation. What this means is that it'll raise an exception that provides us some detail of where we went wrong.

```python
go.Scatter(markers={'color':'blue'})
```

Based on the error log `markers` is not a key in `scatter`. Instead as we browse the list, we see that what we wanted was actually `marker` (singular, without the s), which is a keyword. Thus this little feature helps with easier debugging and correction.


Now let's talk about the methods that come along with these objects. The one of most importance would be the `update` method. The difference here between the regular update method for dictionaries is that it allows for `nested updates`. 
Let me show you what that means.

```python
scatter_trace = go.Scatter(marker={'color':'blue'})
```

We add some information like x and y, as well as updating the sybmol we want the markers to be

```python
scatter_trace.update({'x':[1,2,3],'y':[4,5,6], 'marker':{'symbol':'star', 'size':10, 'color':'#6B48FF'}})
scatter_trace
```

```python
figure = go.Figure(data=[scatter_trace], layout=layout)
py.iplot(figure, filename='pyguide_2')
```

Notice that we were able to add the data as well as add the extra feature for the marker. However if we let `scatter_trace` be just a standard dictionary then we will not be able to just add another feature for the marker in this way. Instead, the value for marker will be replaced with the dictionary we choose to update with.

```python
scatter_trace = {'marker': {'color': 'blue'},
                 'type': 'scatter',
                 ''}
```

As an example, say we looked at two different types of plots in the `heatmap` and the `box plots`, now equipped with the knowledge of subplots, we can easily two and two together in order to put both these on the same plotting surface (for no other reason than that we can). Let's first load the packages, and then set up the data for our heatmap trace.

```python
from plotly import tools
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go

heatmap = go.Heatmap(z=[[1, 20, 30], [20, 1, 60], [30, 60, 1]], showscale=False)
```

Next, we'll set up the trace object for our wind rose chart, and note I'm just copying code from the example pages for each of these plot types.

```python
y0 = np.random.randn(50)
y1 = np.random.randn(50)+1

box_1 = go.Box(y=y0)
box_2 = go.Box(y=y1)
data = [heatmap, box_1, box_2]
```

```python
fig = tools.make_subplots(rows=2, cols=2, specs=[[{}, {}], [{'colspan': 2}, None]],
                          subplot_titles=('First Subplot', 'Second Subplot', 'Third Subplot'))

fig.append_trace(box_1, 1, 1)
fig.append_trace(box_2, 1, 2)
fig.append_trace(heatmap, 2, 1)

fig['layout'].update(height=600, width=600, title='i <3 subplots')

py.iplot(fig, filename='box_heatmap1')
```

This looks great, however we'd like for our subplots to be on the same plotting surface. So let's take a look at the dictionary representation of our figure, and customize it using what we've learned before. 

We actually see that the second boxplot has its own xaxis and yaxis. Thus we should unify it with the same axis as the first subplot. Then from the layout section, we should remove the additional xaxis and yaxis that are drawn for us. You can perform these steps seperately to see what I mean, but in this guide, I'll just show you the result of this edit/customization.

```python
fig.data[1].yaxis = 'y1'
fig.data[1].xaxis = 'x1'
fig.layout.yaxis2 = {}
fig.layout.xaxis2 = {}
```

Now we still have to remove the annotation for the `Second Subplot` title, asell as the `First Subplot` title, and then extend the range of xaxis1 to the entire plotting surface.

```python
fig.layout.annotations = [fig.layout.annotations[2]]  #deletes annotation for the `First and Second Subplot`
fig.layout.xaxis1.domain = [0.0, 1]
```

Voila! We just put together some examples that we've never seen before, and customized out plot using what we learned throughout the guide!

```python
py.iplot(fig, filename='box-heatmap-fixed')
```
