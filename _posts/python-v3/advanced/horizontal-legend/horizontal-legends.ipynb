###  Horizontal Legend

```python
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np

trace1 = go.Scatter(
                x=np.random.randn(75),
                mode='markers',
                name="Plot1",
                marker=dict(
                size=16,
                color='rgba(152, 0, 0, .8)'
                ))
trace2 = go.Scatter(
                x=np.random.randn(75), 
                mode='markers',
                name="Plot2",
                marker=dict(
                size=16,
                color='rgba(0, 152, 0, .8)'
                ))
trace3 = go.Scatter(
                x=np.random.randn(75), 
                mode='markers',
                name="Plot3",
                marker=dict(
                size=16,
                color='rgba(0, 0, 152, .8)'
                ))

data = [trace1, trace2, trace3]
layout = go.Layout(
                legend=dict(
                orientation="h")
                )
figure=go.Figure(data=data, layout=layout)

py.iplot(figure)
```
